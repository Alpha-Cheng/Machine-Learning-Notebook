```python
import os
import numpy as np
import torch
import random
from torch.utils.data import DataLoader
import torchvision.transforms as transforms
from tools.my_dataset import RMBDataset
from PIL import Image
from matplotlib import pyplot as plt
```

```python
def set_seed(seed=1):
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    torch.cuda.manual_seed(seed)


set_seed(1)  # 设置随机种子

# 参数设置
MAX_EPOCH = 10
BATCH_SIZE = 1
LR = 0.01
log_interval = 10
val_interval = 1
rmb_label = {"1": 0, "100": 1}


def transform_invert(img_, transform_train):
    """
    将data 进行反transfrom操作
    :param img_: tensor
    :param transform_train: torchvision.transforms
    :return: PIL image
    """
    if 'Normalize' in str(transform_train):
        norm_transform = list(filter(lambda x: isinstance(x, transforms.Normalize), transform_train.transforms))
        mean = torch.tensor(norm_transform[0].mean, dtype=img_.dtype, device=img_.device)
        std = torch.tensor(norm_transform[0].std, dtype=img_.dtype, device=img_.device)
        img_.mul_(std[:, None, None]).add_(mean[:, None, None])

    img_ = img_.transpose(0, 2).transpose(0, 1)  # C*H*W --> H*W*C
    img_ = np.array(img_) * 255

    if img_.shape[2] == 3:
        img_ = Image.fromarray(img_.astype('uint8')).convert('RGB')
    elif img_.shape[2] == 1:
        img_ = Image.fromarray(img_.astype('uint8').squeeze())
    else:
        raise Exception("Invalid img shape, expected 1 or 3 in axis 2, but got {}!".format(img_.shape[2]) )

    return img_


# ============================ step 1/5 数据 ============================
train_dir = os.path.join("test_data")

norm_mean = [0.485, 0.456, 0.406]
norm_std = [0.229, 0.224, 0.225]
```

```python
train_transform = transforms.Compose([
transforms.Resize((224, 224)),
transforms.ToTensor(),
transforms.Normalize(norm_mean, norm_std),
])
def show_img(train_transform):
    
    # 构建MyDataset实例
    train_data = RMBDataset(data_dir=train_dir, transform=train_transform)

    # 构建DataLoder
    train_loader = DataLoader(dataset=train_data, batch_size=BATCH_SIZE)
    for i, data in enumerate(train_loader):

        inputs, labels = data   # B C H W

        img_tensor = inputs[0, ...]     # C H W
        img = transform_invert(img_tensor, train_transform)    
        plt.imshow(img)
        plt.show()
```

## 原始图片

```python
show_img(train_transform)
```

# 一. 裁剪 -----Crop

## 1. CenterCrop(size)   从中间裁剪

+ size比原图片小

```python
train_transform = transforms.Compose([
transforms.Resize((224, 224)),
transforms.CenterCrop(196),
transforms.ToTensor(),
transforms.Normalize(norm_mean, norm_std),
])
show_img(train_transform)
```

+ size比图片大时自动填充

```python
train_transform = transforms.Compose([
transforms.Resize((224, 224)),
transforms.CenterCrop(512),
transforms.ToTensor(),
transforms.Normalize(norm_mean, norm_std),
])
show_img(train_transform)
```

## RandomCrop(size,padding,pad_if_need,fill,padding_mod)
## 从图片中随机裁剪
padding_mode ---填充模式
+ constant: 像素值由fill设定
+ edge: 像素值由图像边缘像素决定
+ reflect: 镜像填充
+ symmetric: 最后一个像素也镜像

```python
train_transform = transforms.Compose([
transforms.Resize((224, 224)),
transforms.RandomCrop(150,padding=16),
transforms.ToTensor(),
transforms.Normalize(norm_mean, norm_std),
])
show_img(train_transform)
```

```python
train_transform = transforms.Compose([
transforms.Resize((224, 224)),
transforms.RandomCrop(224,padding=16,fill=133),
transforms.ToTensor(),
transforms.Normalize(norm_mean, norm_std),
])
show_img(train_transform)
```

```python
train_transform = transforms.Compose([
transforms.Resize((224, 224)),
# size大于图像尺寸是需要设置pad_if_need
transforms.RandomCrop(361,padding=64,pad_if_needed=True,padding_mode='edge'),  
transforms.ToTensor(),
transforms.Normalize(norm_mean, norm_std),
])
show_img(train_transform)
```

```python
train_transform = transforms.Compose([
transforms.Resize((224, 224)),
transforms.RandomCrop(1024,padding=64,pad_if_needed=True,padding_mode='reflect'),
transforms.ToTensor(),
transforms.Normalize(norm_mean, norm_std),
])
show_img(train_transform)
```

## RandomResizedCrop(size,scale,ratio,interpolation)
+ size: 所需裁剪的图片尺寸
+ scale: 随机裁剪面积比例
+ ratio: 随机长宽比

```python
train_transform = transforms.Compose([
transforms.Resize((224, 224)),
transforms.RandomResizedCrop(152,scale=(0.5,1),ratio=(3/4,4/3)),
transforms.ToTensor(),
transforms.Normalize(norm_mean, norm_std),
])
show_img(train_transform)
```

## FiveCrop(size) 
-----左上左下,右上右下,中间裁剪出五张图片
## TenCrop(size,vertica_flip)   
-----对FiveCrop裁剪出来的五张图片分别取镜像对称

```python
train_transform = transforms.Compose([
transforms.Resize((224, 224)),

# FiveCrop返回的是Tuple需要使用lambda函数,不需要toTensor和Normalize
transforms.FiveCrop(112),
transforms.Lambda(lambda crops: torch.stack([(transforms.ToTensor()(crop)) for crop in crops])),

# transforms.ToTensor(),
# transforms.Normalize(norm_mean, norm_std),
])

# 构建MyDataset实例
train_data = RMBDataset(data_dir=train_dir, transform=train_transform)

# 构建DataLoder
train_loader = DataLoader(dataset=train_data, batch_size=BATCH_SIZE)

#展示图片你需要每个维度都可视化
for epoch in range(1):
    for i, data in enumerate(train_loader):

        inputs, labels = data   # B C H W

        bs, ncrops, c, h, w = inputs.shape
        for n in range(ncrops):
            img_tensor = inputs[0, n, ...]  # C H W
            img = transform_invert(img_tensor, train_transform)
            plt.imshow(img)
            plt.show()
            plt.pause(1)
```

```python
train_transform = transforms.Compose([
transforms.Resize((224, 224)),

# FiveCrop返回的是Tuple需要使用lambda函数,不需要toTensor和Normalize
transforms.TenCrop(112,vertical_flip=True),    # 垂直翻转
transforms.Lambda(lambda crops: torch.stack([(transforms.ToTensor()(crop)) for crop in crops])),

# transforms.ToTensor(),
# transforms.Normalize(norm_mean, norm_std),
])

# 构建MyDataset实例
train_data = RMBDataset(data_dir=train_dir, transform=train_transform)

# 构建DataLoder
train_loader = DataLoader(dataset=train_data, batch_size=BATCH_SIZE)

#展示图片你需要每个维度都可视化
for epoch in range(1):
    for i, data in enumerate(train_loader):

        inputs, labels = data   # B C H W

        bs, ncrops, c, h, w = inputs.shape
        for n in range(ncrops):
            img_tensor = inputs[0, n, ...]  # C H W
            img = transform_invert(img_tensor, train_transform)
            plt.imshow(img)
            plt.show()
            plt.pause(1)
```

# 二. 翻转   ----- Flip

## RandomHorizontalFlip(p)         ----水平翻转
## RandomVerticalFlip(p)          ----垂直反转

```python
train_transform = transforms.Compose([
transforms.Resize((224, 224)),
transforms.RandomHorizontalFlip(0.5),
transforms.RandomVerticalFlip(0.5),
transforms.ToTensor(),
transforms.Normalize(norm_mean, norm_std),
])
show_img(train_transform)
```

# 旋转    -----Rotation

## RandomRotation(degrees,resample,expand,center)
+ degrees : 旋转角度,输入a时在(-a,a)之间旋转,(a,b)时在(a,b)旋转
+ resample : 重采样方法
+ expand : 是否扩大图片,以保持原图信息(超出矩形框部分)
+ center : 旋转中心

```python
train_transform = transforms.Compose([
transforms.Resize((224, 224)),
transforms.RandomRotation(90),
transforms.ToTensor(),
transforms.Normalize(norm_mean, norm_std),
])
show_img(train_transform)
```

```python
train_transform = transforms.Compose([
transforms.Resize((224, 224)),
transforms.RandomRotation(90,expand=True),
transforms.ToTensor(),
transforms.Normalize(norm_mean, norm_std),
])
show_img(train_transform)
```

```python
train_transform = transforms.Compose([
transforms.Resize((224, 224)),
# expand并不能不全丢失信息
transforms.RandomRotation(90,expand=True,center=(0,0)),
transforms.ToTensor(),
transforms.Normalize(norm_mean, norm_std),
])
show_img(train_transform)
```
